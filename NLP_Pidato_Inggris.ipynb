```python
import requests
import urllib.request
#response =  urllib.request.urlopen('https://en.wikipedia.org/wiki/2022_State_of_the_Union_Address')
response =  urllib.request.urlopen('https://www.whitehouse.gov/state-of-the-union-2022/')
html = response.read()
print(html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html.parser')
text = soup.get_text(strip = True)
print(text)
```

```python
tokens = [t for t in text.split()]
print(tokens)
```

```python
from nltk.corpus import stopwords
sr= stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        
        clean_tokens.remove(token)
```

```python
import nltk
nltk.download('punkt')  # Mengunduh data yang diperlukan oleh NLTK

# Kemudian Anda dapat melanjutkan dengan mengimpor pustaka atau modul yang diperlukan lainnya
# Misalnya, jika Anda menggunakan nltk.tokenize atau nltk.corpus, Anda juga perlu mengimpor mereka.

# Selanjutnya, Anda dapat menggunakan kode Anda seperti yang telah Anda tulis sebelumnya.
freq = nltk.FreqDist(clean_tokens)
for key, val in freq.items():
    print(str(key) + ':' + str(val))
```

```python
freq.plot(20, cumulative=False)
```
