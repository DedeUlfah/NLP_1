```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from collections import Counter

# Baca teks pidato dari file
with open('C:\\Users\\Asus\\Downloads\\nlp_ulfah\\pidato-kenegaraan-presiden.txt', 'r', encoding='utf-8') as file:
    pidato = file.read()

# Tokenisasi teks
tokens = word_tokenize(pidato)

# Hapus kata-kata stopwords (kata-kata umum yang tidak memberikan makna khusus)
stop_words = set(stopwords.words('indonesian'))
filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]

# Hitung frekuensi kata-kata
word_freq = Counter(filtered_tokens)

# Ambil 20 kata yang paling sering muncul
common_words = word_freq.most_common(20)

# Siapkan data untuk plot
words, frequencies = zip(*common_words)

# Plot kata-kata yang paling sering muncul
plt.figure(figsize=(12, 6))
plt.bar(words, frequencies)
plt.xlabel('Kata-kata')
plt.ylabel('Frekuensi')
plt.title('20 Kata Paling Sering Muncul dalam Pidato')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
